```python
#Explorando Dados e Usando Gráficos
# Usando Matplotlib para desenhar um histogram para a coluna idade.
plt.hist(df['Age'],bins=8)
```

```python
# Usando o seaborn package para desenhar o histograma
sns.distplot(df['Age']);
```

```python
# O uso regular do matplotlib para mostrar o "Gráfico de Pizza" barplot
df.groupby(['Sex'])['Age'].mean().plot(kind='pie')
```

```python
# O uso do seaborn para mostrar o Gráfico de Barras
sns.set_style("darkgrid")
ax = sns.barplot(x='Sex',y ='Age', data=df, estimator=len)
```

```python
# Dividindo o gráfico em 02 grupos:
ax = sns.barplot(x='Sex',y ='Age', hue='Embarked', data=df, estimator=len)
```

```python
#Violinplot
sns.violinplot(x = "Age", data=df)
```

```python
#Scatterplot in seaborn
sns.jointplot(x='Parch', y='Age', data=df)
```

```python
#Se estivermos em gráficos para representar a regressão linear para 2 variáveis numéricas - regplot
sns.regplot(x='Parch', y='Age', data=df)
```

```python
# box plot - é um tipo de Gráfico muito utilizado para detectar Out-lier
sns.boxplot(x='Survived',y='Age', hue='Sex', data=df)
```

```python
# side-by-side box plot - com parâmetros de Rank com idade aplicados ao Sexo
sns.boxplot(x='Survived',y='Age', data=df, hue='Sex')
```

```python
#catplot - Gráficos de Barras do tipo: Catplot
sns.catplot(x='Pclass',y='Fare', data=titanic, kind='bar')
```
